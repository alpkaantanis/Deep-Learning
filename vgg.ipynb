```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense,Dropout
from tensorflow.keras.layers import Conv2D,MaxPooling2D,ZeroPadding2D
```

```python
model=Sequential()
##blok-1
model.add(Conv2D(input_shape=(224,224,3),filters=64,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(64,kernel_size=(3,3),padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
##blok-2
model.add(Conv2D(128,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(128,kernel_size=(3,3),padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
##blok-3
model.add(Conv2D(256,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(256,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(256,kernel_size=(3,3),padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
##blok-4
model.add(Conv2D(512,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(512,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(512,kernel_size=(3,3),padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))
##blok-5
model.add(Conv2D(512,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(512,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Conv2D(512,kernel_size=(3,3),padding='same',activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=(2,2)))

##fully connected layers
model.add(Flatten())
model.add(Dense(4096,activation='relu'))
model.add(Dense(4096,activation='relu'))
model.add(Dense(1000,activation='softmax'))
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import load_img,img_to_array
image=load_img('/Ew8GRzzWgAAoWCe.jpg',target_size=(224,224))
image=img_to_array(image)
image=image.reshape((1,image.shape[0],image.shape[1],image.shape[2]))

from tensorflow.keras.applications.vgg16 import VGG16

Vgg16Model=VGG16(  include_top=True,weights="imagenet")

from tensorflow.keras.applications.vgg16 import preprocess_input,decode_predictions

image=preprocess_input(image)

pred=Vgg16Model.predict(image)

label=decode_predictions(pred)

print(label)
```

```python
model=Sequential()
model.add(Dense(10,input_shape=(8,),activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.summary()

```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input

inp=Input(shape=(8,))

hidden1=Dense(10,activation='relu')(inp)
hidden2=Dense(20,activation='relu')(hidden1)
hidden3=Dense(10,activation='relu')(hidden2)
out=Dense(1,activation='sigmoid')(hidden3)

mymodel=Model(inputs=inp,outputs=out)
mymodel.summary()
```

```python
from tensorflow.keras.applications import VGG16
vgg=Sequential()
vgg.add(VGG16(input_shape=(224,224,3),weights='imagenet',include_top=False))
vgg.add(Flatten())
vgg.add(Dense(512,activation='relu'))
vgg.add(Dense(6,activation='softmax'))

vgg.summary()
```

```python
base_model=VGG16(input_shape=(224,224,3),include_top=False)
x=Flatten()(base_model.output)
x=Dense(512,activation='relu')(x)
out=Dense(6,activation='softmax')(x)
model=Model(inputs=base_model.inputs,outputs=out)
model.summary()

```
